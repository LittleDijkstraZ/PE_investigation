```python
def get_sample_num_from_file_df(file):
    df = pd.read_csv(file)
    # make additional column for number of samples by extracting from 'model_name' with splitting with '_' and selecting -2-th element
    df['num_train_samples'] = df['model_name'].apply(lambda x: int(x.split('_')[-2]))
    df.sort_values(by=['num_train_samples'], inplace=True)
    
    return df
```

```python
def get_num_n_plus_1(n=3, num_sample=10000):
    d = int((num_sample-100)  / (n*(n+1)/2 - 1))
    p_list = [100] + [i*d for i in range(2, n+1)]
    if p_list[1] > 10000:
        p_list[1] = 10000
        p_list[2] += num_sample - sum(p_list)
    else:
        p_list[1] += num_sample - sum(p_list)
    return p_list[-1]


def get_new_sample_num_from_file_df(file, digit):
    df = pd.read_csv(file)
    # make additional column for number of samples by extracting from 'model_name' with splitting with '_' and selecting -2-th element
    df['num_train_samples'] = df['model_name'].apply(lambda x: int(x.split('_')[-2]))
    df['new_num_train_samples'] = df['num_train_samples'].apply(lambda x: get_num_n_plus_1(n=digit, num_sample=x))
    df.sort_values(by=['num_train_samples'], inplace=True)
    
    return df
```

# n-digit -> n+1 digit

```python
def get_num_n_plus_1(n=3, num_sample=10000):
    d = int((num_sample-100)  / (n*(n+1)/2 - 1))
    p_list = [100] + [i*d for i in range(2, n+1)]
    if p_list[1] > 10000:
        p_list[1] = 10000
        p_list[2] += num_sample - sum(p_list)
    else:
        p_list[1] += num_sample - sum(p_list)
    return p_list[-1]


def get_new_sample_num_from_file_df(file, digit):
    df = pd.read_csv(file)
    # make additional column for number of samples by extracting from 'model_name' with splitting with '_' and selecting -2-th element
    df['num_train_samples'] = df['model_name'].apply(lambda x: int(x.split('_')[-2]))
    df['new_num_train_samples'] = df['num_train_samples'].apply(lambda x: get_num_n_plus_1(n=digit, num_sample=x))
    df.sort_values(by=['num_train_samples'], inplace=True)
    
    return df
```
