```python
import pandas as pd
import numpy as np
import math

import matplotlib.pyplot as plt
```

```python
def truncate_to_4_digit(x):
    return math.floor(x * 10000) / 10000
```

```python

metrics = ['accuracy_eps0','accuracy_eps5e-4']
labels = ['plain, eps=0', 'ar, eps=0', 'plain, eps=5e-4', 'ar, eps=5e-4']
colors = ['#0057a8','#f83b36','#17becf','#ff7f0e']
# ['#1f77b4', ''#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
# ['#0057a8', '#f83b36', '#79b84e', '#ff80ed', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#f83b36']
files = ['out2/sin/eval_test_sin_overall/error_df.csv','out2/sin_ar/eval_test_sin_ar_overall/error_df.csv' ]

i = 0
plt.figure(figsize=(8,6))
for metric in metrics:
    for filename in files:
        df = pd.read_csv(filename)

        bins = np.arange(-100,100,2)/100
        # df['bin'] = pd.cut(df['y'], bins)
        group = df.groupby(pd.cut(df.y, bins))
        plot_centers = (bins [:-1] + bins [1:])/2
        plot_values = group[metric].mean()

        plt.plot(plot_centers, plot_values, label=labels[i], color=colors[i], linewidth=4, markersize=12)
        i+=1

plt.title("Sin", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
# plt.ylim(0, 100)
plt.xlabel('true y value',fontsize=24)
plt.ylabel('Test Accuracy (%)',fontsize=24)
plt.legend(loc='lower right', fontsize=16)
plt.grid(True)
plt.tight_layout()

plt.savefig(f'figs/sin_accuracy.pdf', format='pdf', dpi=600, bbox_inches='tight', pad_inches=0.0)
```
